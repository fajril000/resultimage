```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca citra (ganti 'gambar.jpg' dengan nama file citra Anda)
img = cv2.imread('download.jpeg', 0)

# Hitung histogram
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# Tampilkan citra asli dan histogramnya dalam satu figure
plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.subplot(1,2,2), plt.plot(hist)
plt.title('Histogram Citra')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Pixel')
plt.xlim([0, 256])
plt.show()
```

```python
hist,bins = np.histogram(img.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()
```

```python
cdf_m = np.ma.masked_equal(cdf,0,copy=True)
cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
cdf = np.ma.filled(cdf_m,0).astype('uint8')
```

```python
img_equalized = cdf[img]
```

```python
plt.subplot(121),plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.subplot(122),plt.imshow(img_equalized, cmap='gray')
plt.title('Citra Setelah Equalization')
plt.show()
```

```python
hist_equalized, _ = np.histogram(img_equalized.flatten(), 256, [0, 256])
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Baca gambar grayscale
img = cv2.imread('download.jpeg', 0)

# Histogram Equalization
equ = cv2.equalizeHist(img)

# Hitung histogram gambar asli, hasil equalization
hist_img = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_equ = cv2.calcHist([equ], [0], None, [256], [0, 256])

# Tampilkan gambar asli, equalization dan histogram masing-masing
plt.figure(figsize=(15, 12))

# Gambar Asli
plt.subplot(4, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(4, 2, 2)
plt.plot(hist_img, color='black')
plt.title('Histogram Gambar Asli')

# Hasil Equalization
plt.subplot(4, 2, 3)
plt.imshow(equ, cmap='gray')
plt.title('Hasil Equalization')
plt.axis('off')

plt.subplot(4, 2, 4)
plt.plot(hist_equ, color='black')
plt.title('Histogram Hasil Equalization')
plt.tight_layout()
plt.show()

```
